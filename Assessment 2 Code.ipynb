```python
#4.4.3 load and read CSV files, add the pandas dependancy 
import pandas as pd 

state_data_to_load= "Assign_2/donors2008.csv"

#read the school data file and store it in a Pandas Dataframe 
state_data_df=pd.read_csv(state_data_to_load)
state_data_df
```

```python
#we dont want all that data in the summary df, so to get the average math score and reading score we can add the columns at the end 
#calc the average test scores
per_state_amount=state_data_df.groupby(["State"]).mean()["Amount"]
per_state_amount
```

```python
per_state_summary_df=pd.DataFrame({
    "State": per_state_amount,
    })
per_state_summary_df
```

```python
greater_750=per_state_summary_df[per_state_summary_df["State"]>=750]
greater_750
```
